```python
%matplotlib inline
from imageio import imread, imwrite
import matplotlib.pyplot as plt
import align.detect_face
import tensorflow as tf
from tqdm import tqdm
import pandas as pd
import numpy as np
import cv2
import os
```

```python
path_to_imgs = '/home/facialrec/notebooks/datasets/VGGFace2/test/'
path_to_labels = '/home/facialrec/notebooks/datasets/VGGFace2/bb_landmark/loose_bb_test.csv'
target_folder = '/home/facialrec/notebooks/datasets/VGGFace2/test_cropped/'

df = pd.read_csv(path_to_labels)
```

```python
for idx, record in tqdm(df.iterrows()):
#         print(record['NAME_ID'], record['X'], record['Y'], record['W'], record['H'])
        img_path = os.path.join(path_to_imgs, record['NAME_ID']+'.jpg')
        img = imread(img_path)
        y = max(record['Y'], 0)
        y2 = max(record['Y']+record['H'], 0)
        x = max(record['X'], 0)
        x2 = max(record['X']+record['W'], 0)
        face_img = img[y:y2, x:x2]
        
        index = record['NAME_ID'].find('/')
        folder = record['NAME_ID'][:index]
        folder_path = os.path.join(target_folder, folder)
        os.makedirs(folder_path, exist_ok=True)
        cropped_img_path = os.path.join(target_folder, record['NAME_ID']+'.jpg')
        imwrite(cropped_img_path, face_img)
```
